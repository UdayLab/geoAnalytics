# **Operatioin 1: Raster to CSV**

## ***Step1: Import RasterToTSV package from geoAnalytics.conversion***

```python
from geoanalytics.conversion import Raster2CSV
```

## ***Step2: Pass the lbl file as input and give desired outputFile name also specify the inputband value as well as ouput band value***

```python
converter = Raster2CSV.Raster2CSV(inputFile='MI_MAP_03_S79E042S80E043SC.lbl', outputFile='Moon.csv', startBand=1, endBand=9)
```

## ***Step3: Conver the raster file into a CSV file***

```python
converter.run()
```

# **Operation 2: Imputing the Data**

## ***Step 1: Read the obtained CSV into a dataframe***

```python
import numpy as np
import pandas as pd
df = pd.read_csv('Moon.csv', sep='\t')
df
```

### **Since in a raster file, Nan values will not be stored. We assume that zeros are the missing data**

```python
print(df.isna().sum().sum())
```

```python
print((df == 0).sum().sum())
```

## ***Step 2: Replace the zeros with Nan***

```python
df.replace(0, np.nan, inplace=True)
```

```python
print(df.isna().sum().sum())
```

## ***Step 3: Import any Imputation algorithm from geoanalytics.imputation***

```python
from geoanalytics.imputation import MeanImputation
obj = MeanImputation.MeanImputation(dataframe=df)
```

## ***Step 4: Execute the run() method to impute the data***

### cross-check once again by printing the Nan values in resultat dataframe

```python
ImputedDF = obj.run()
print(ImputedDF.isna().sum().sum())
```

## ***Step5: To know the run time and memory consumption of Algorithm, print the statistical values***

```python
obj.getRuntime()
obj.getMemoryRSS()
obj.getMemoryUSS()
```

## ***Step6: Save the labels and centers***

```python
obj.save(outputFile='MeanImputation.csv')
```

# **Operation3: CSV to Raster**

## ***Step1: Import CSV2Raster from geoanalytics.conversion***

```python
from geoanalytics.conversion import CSV2Raster as CSV2Raster
```

## ***Step2: Pass the dataFrame and give desired outputFile name***

```python
process = CSV2Raster.CSV2Raster(dataframe=ImputedDF,outputFile='MeanImputation.tiff')
```

## ***Step3: Execute the run() method***

```python
process.run()
```
